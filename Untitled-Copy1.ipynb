```python
import IPython.display
import pandas as pd
import os
from czsc.data.ts_cache import TsDataCache
from examples.collect_datas.run_thsplates_monitor import getPlatesRankDf, calculatePlateInsideStock
from examples.run_allplates_monitor import CzscStocksPlates
from czsc.utils.io import read_pkl
import copy

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

ct_path = "/Users/guyeqi/Documents/Python/data/realtime/platesdata"
relations =pd.read_csv(os.path.join(ct_path, "relations"), index_col=0)
cachedMembers = read_pkl(os.path.join(ct_path, "members"))

#股票 -> 概念
def stockToPlates(stock):
    return relations[relations[stock] == 1].index.tolist()

#概念 -> 股票
def plateToStocks(plate):
    relations2 = relations.T
    plateDF = relations2.loc[:, ~relations2.columns.duplicated()]
    return plateDF[plateDF[plate] == 1].index.tolist()

# 获取过去【N】天涨跌幅
# code zhName SMAX1 SMAX2....
templateRankInfo = getPlatesRankDf()
display(templateRankInfo)

def getInfo(day, yuzhi):
    #{概念1：[成分股组1]，概念2：[成分股组2].....}
    eachPlateStocks = {}
    col = ["标的", "出现次数", "所属概念"]
    dfEachPlateStocks = pd.DataFrame(columns=col)
#     dfEachPlateStocks = pd.concat([dfEachPlateStocks, pd.DataFrame.from_records([result])], ignore_index=True)
    #根据 key 排序
    head = 20
    key = "SMA"+str(day)
    sortedInfo = templateRankInfo.sort_values(key, inplace=False, ascending=False)
    sortedInfo= sortedInfo.loc[:,["code", "zhName", key]]
    sortedInfo = pd.concat([sortedInfo, pd.DataFrame(columns=['成分股'])], sort=False)
    headInfo = sortedInfo.head(n=head).reset_index(drop=True)
    # display(sortedInfo.tail(n=10).reset_index(drop=True))
    for index, row in headInfo.iterrows():
        print("正在计算成分股涨跌幅 【" + row["zhName"] + "】")
        eachPlateStocks[row["zhName"]] = plateToStocks(row["zhName"])
        stockInfoResult = calculatePlateInsideStock(day, row["zhName"], yuzhi)
        headInfo.loc[headInfo["zhName"]==row["zhName"], "成分股"]= str(stockInfoResult)
    display(headInfo)
    
    hasCalStocks = []
    eachPlateStocksCopy = copy.deepcopy(eachPlateStocks)
    for key,value in eachPlateStocks.items():
        #key 概念，value 该概念所有股票
        for oneStockZhName in value:
            plateForThisOne = []
            #统计过不需要统计
            if oneStockZhName in hasCalStocks:
                continue
            #开始重新遍历
            hasCalStocks.append(oneStockZhName)
            for key,value in eachPlateStocksCopy.items():
                if oneStockZhName in value:
                    plateForThisOne.append(key)
            #统计完成，df合并
            oneStockResult = {"标的":oneStockZhName, "出现次数": len(plateForThisOne), "所属概念":str(plateForThisOne)}
            dfEachPlateStocks = pd.concat([dfEachPlateStocks, pd.DataFrame.from_records([oneStockResult])], ignore_index=True)
    dfEachPlateStocks = dfEachPlateStocks.loc[dfEachPlateStocks['出现次数'] > 2]
    dfEachPlateStocks2 = dfEachPlateStocks.drop_duplicates()
    display(dfEachPlateStocks2.sort_values("出现次数", inplace=False, ascending=False))
```
